```python
words=["eating","eats", "eaten", "working", "worked", "writes", "programming"]
```

```python
from nltk.stem import PorterStemmer

stemming=PorterStemmer()

for word in words: 
    print(word+ "--->" + stemming.stem(word))
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)

reg_stemmer.stem("eating")
reg_stemmer.stem("understandable")
```

```python
reg_stemmer.stem("writing")
```

```python
### Snowball stemer -> it gives better accuracy

from nltk.stem import SnowballStemmer
snowballsstemer=SnowballStemmer("english")
```

```python
for word in words:
    print(words+ "---->"+ snowballsstemer.stem(word) ) 
```
